```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
faculty = pd.read_csv("../data/faculty.csv")
faculty["last_name"] = [f.split(',')[0] for f in faculty["name"]]
faculty["first_middle_name"] = [f.split(',')[1].lstrip() for f in faculty["name"]]
```

```python
faculty.groupby(['dept'])['salary'].min()
faculty.groupby(['dept'])['salary'].mean()
faculty.groupby(['dept'])['salary'].median()
faculty.groupby(['dept'])['salary'].max()
```
